```python
# Importing all the required librairies
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb
```

```python
# Defining the array
A = np.array([[2, -1], [-1, 2]])
B = np.array([0, 3])
```

```python
# Numpy solver
x = np.linalg.solve(A, B)
print(x)
```

```python
# Matrix inverse
A_inv = np.linalg.inv(A)
x_inv = np.dot(A_inv, B)
print(x_inv)
```

```python
# Calculating linear combination
VA1 = A[:, 0]
VA2 = A[:, 1]
scaled_A1 = VA1 * x[0]
scaled_A2 = VA2 * x[1]
linear_combination = scaled_A1 + scaled_A2
print(linear_combination)
```

```python
# Visualising the plot
plt.figure(figsize=(8, 6))

plt.plot([0, VA1[0]], [0, VA1[1]], label='VA1')
plt.plot([0, VA2[0]], [0, VA2[1]], label='VA2')
plt.plot([0, scaled_A1[0]], [0, scaled_A1[1]], label='VA1 scaled by X[0]')
plt.plot([scaled_A1[0], scaled_A1[0]+ scaled_A2[0]], [scaled_A1[1], scaled_A1[1] + scaled_A2[1]], label='VA2 scaled by X[1]')
plt.plot([0, linear_combination[0]], [0, linear_combination[1]], label='Linear combination')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Linear Combination of Column Vectors")
plt.legend()
plt.grid(True)
plt.show()
```
