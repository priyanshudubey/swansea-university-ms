```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import comb

n = 100
p = 0.7
num_trials = 12000

num_success = np.zeros(num_trials, dtype=int)
print(num_success)

for i in range(num_trials):
    success = np.random.choice([0, 1], size=n, p=[1-p, p])
    num_success[i] = np.sum(success)

plt.figure(figsize=(8, 6))
plt.hist(num_success, bins=range(n+1), alpha=0.5, color='blue')

x = np.arange(n+1)
y = np.zeros(n+1)
for i in range(n+1):
    y[i] = comb(n, i) * (p ** i) * ((1-p) ** (n-i))
# plt.plot(x, y, "k-", label="Theory")

plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title("Binomial Distribution")
plt.show()
```
