```python
# Importing important liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# Loading the data
x_values = np.load("x_values-1.npy")
print(x_values)
y_values = np.load("y_values-1.npy")
print(y_values)
```

### Task 3.1

```python
plt.scatter(x_values,y_values, label = "values")
plt.title("Scatter plot of x_values and y_values")
plt.xlabel("Regular Variable")
plt.ylabel("Target Variable")
plt.legend()
plt.show()
```

```python
# Create linear regression object and fit the model
model = LinearRegression()
reg = model.fit(x_values, y_values)
reg.score(x_values, y_values)
# Predict the y_values
predicted_y = reg.predict(x_values)
```

```python
# Plotting
plt.scatter(x_values,y_values, label = "training data")
plt.plot(x_values, predicted_y, color = "red", label = "Predicted data")
plt.title("Linear Regression")
plt.xlabel("Regular Variable")
plt.ylabel("Target Variable")
plt.legend()
plt.show()
```

```python
# For printing the y_value for x = 0.48
x = [[0.48]]
y = reg.predict(x)
print("Printed the value of y for x = 0.48 : ", y)
```

### Task 3.2

```python
# Loading wine data
wData = np.load("wineData-1.npy")
wLabel = np.load("wineLabels-1.npy")
```

```python
print(wData)
print(wLabel)
```

```python
# splitting the data and labels into test and train set
train_data, test_data, train_labels, test_labels = train_test_split(wData, wLabel, test_size = 0.2, random_state = 42)
pca = PCA(n_components = 0.8)
train_data_pca = pca.fit_transform(train_data)
test_data_pca = pca.transform(test_data)
# print(train_data_pca)
```

```python
# plot
plt.scatter(train_data[:, 0], train_data[:, 1], c = train_labels, marker = 'o', label='Training Data')
plt.scatter(test_data[:, 0], test_data[:, 1], c = test_labels, marker = 'x', label = 'Test Data')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.title('Plotting the training and test data againsta feature')
plt.legend()
plt.show()
```

```python
# Reducing dimensionality
pca = PCA(n_components = 2)
train_data_pca = pca.fit_transform(train_data)
# print(train_data_pca[:,1])
test_data_pca = pca.transform(test_data)
```

```python
# plot
plt.scatter(train_data_pca[:, 0], train_data_pca[:, 1], c = train_labels, marker = 'o', label='Training')
plt.scatter(test_data_pca[:, 0], test_data_pca[:, 1], c = test_labels, marker = 'x', label = 'Testing')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.title('Plotting the training and test with dimensionality reduction')
plt.legend()
plt.show()
```

## Task 3.3

```python
lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(train_data, train_labels)
```

```python
train_lda = lda.transform(train_data)
test_lda = lda.transform(test_data)
```

```python
plt.scatter(train_data_pca[:, 0], train_data_pca[:, 1], c = train_labels, marker = 'o', label='Training')
plt.scatter(test_data_pca[:, 0], test_data_pca[:, 1], c = test_labels, marker = 'x', label = 'Testing')
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.title('Plotting the training and test with dimensionality reduction')
plt.legend()
plt.show()

plt.scatter(train_lda[:, 0], train_lda[:, 1], c=train_labels, marker='o', label='Training')
plt.scatter(test_lda[:, 0], test_lda[:, 1], c=test_labels, marker='x', label='Testing')

plt.xlabel('Component 0')
plt.ylabel('Component 1')
plt.title('Wine DataSet using LDA')
plt.legend()
plt.show()
```

### Difference seen between PCA and LDA from the above two tasks are :-
### - PCA doesn't care about the different categories or types of data, just about how the data spreads out in all directions.
### - LDA also looks for directions in the data, but it focuses on finding directions that best separate the different categories or classes of data.

## TASK 3.4

```python
m_vector = np.mean(train_data, axis=0)
train_c = train_data - m_vector
U, S, Vt = np.linalg.svd(train_c)
projection_matrix = Vt[:2, :]
train_data_pca2d = train_c @ np.transpose(projection_matrix)
```

```python
test_c = test_data - m_vector
test_data_pca2d = test_c @ np.transpose(projection_matrix)
```

```python
plt.scatter(train_data_pca2d[:, 0], train_data_pca2d[:, 1], c=train_labels, marker='o', label='Training')
plt.scatter(test_data_pca2d[:, 0], test_data_pca2d[:, 1], c=test_labels, marker='x', label='Testing')

plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('wine dataset Projection using PCA')
plt.legend()
plt.show()
```
