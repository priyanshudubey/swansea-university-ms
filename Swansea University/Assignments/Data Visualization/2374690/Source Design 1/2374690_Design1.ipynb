```python
# Importing Liabraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading dataset
df = pd.read_excel('Career Mode female player datasets - FIFA 16-22.xlsx')
print(df.shape)
```

```python
# Checking the shape and type of data
print("Shape of Data set:- ", df.shape)
print("Data type of Data set:- ", df.dtypes)
```

```python
player_positions = df['player_positions'].value_counts().head(20)
player_positions
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=player_positions.index, y=player_positions.values, hue = player_positions.values, palette = 'plasma')
plt.title('Most Common Player Positions')
plt.xticks(fontsize = 15, rotation=90)
plt.yticks(fontsize = 15)
plt.xlabel("Positing on the ground", color = 'red')
plt.ylabel("No of players", color = 'red')
plt.tight_layout()
plt.savefig('design1_img.png')
plt.show()
```
